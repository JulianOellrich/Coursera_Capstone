<h4>3.1.1 Import Munich Borough Data</h4>

```python
# scrape wikipedia page
url = 'https://de.wikipedia.org/wiki/Stadtbezirke_M%C3%BCnchens'
source = requests.get(url).text
soup = BeautifulSoup(source)

# extract data from wikipedia table
table_data = soup.find('div', class_='mw-parser-output')
table = table_data.table.tbody
```

```python
columns = ['Nr.', 'Borough', 'Size in qkm', 'Population', 'Density in Inh/qkm', 'Immigrants']
data = dict({key:[]*len(columns) for key in columns})

for row in table.find_all('tr'):
    for i,column in zip(row.find_all('td'),columns):
        i = i.text
        i = i.replace('\n', '')
        data[column].append(i)

df_MUC_raw = pd.DataFrame.from_dict(data=data)[columns]
```

<h3>4.1 Function definition</h3>

<p>A function <i>getNearbyVenues</i> is created, which extracts all nearby venues in a Borough and puts them into a dataframe an organized way.</p>

<p>The resulting dataframe will have the following columns: 
<ul>
  <li>Borough </li>
  <li>Borough Latitude  </li>
  <li>Borough Longitude  </li>
  <li>Venue </li>
  <li>Venue Latitude </li>
  <li>Venue Longitude  </li>
  <li>Venue Category </li>
</ul>
</p>

```python
def getNearbyVenues(names, latitudes, longitudes, radius = 1500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Borough', 
                  'Borough Latitude', 
                  'Borough Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

<h3>5.2 Clustering boroughs by common venues</h3>

<h4>5.2.1 Function definition</h4>

In this section a number of function will be defined

<h5>A. Function for performing onehot ecoding and grouping</h5>

This function perfons two things:
<ol>
    <li>Perfons one hot encoding on the entered venue dataframe (containing all top venues of each borough) $\Rightarrow$ returns df_onehot</li>
    <li>Groups rows by borough and by taking the mean of the frequency of occurrence of each category $\Rightarrow$ returns df_grouped</li>
</ol>

To run the function on for example the <i>muc_venues</i> dataframe you have to enter the following:

<code> muc_onehot, muc_grouped = explore_boroughs(muc_venues) </code>

Where _muc_onehot_ and _muc_grouped_ contain the mentioned onehot and grouped dataframes

```python
def explore_boroughs(df_venues):
    df_onehot = pd.get_dummies(df_venues[['Venue Category']], prefix ="", prefix_sep ="")
    
    # add Borough column back to dataframe
    df_onehot['Borough'] = df_venues['Borough']

    # move borough column to the first column. 
    # Move last column to the first and perform this 3 times
    fixed_columns = [df_onehot.columns[-1]] + list(df_onehot.columns[:-1])
    df_onehot = df_onehot[fixed_columns]
        
    # Group rows by borough and by taking the mean of the frequency of occurrence of each category
    df_grouped = df_onehot.groupby('Borough').mean().reset_index()
    
    return(df_onehot, df_grouped)
```

<h5>B. Function top extract the top X most common venues of each borough</h5>

The goal is to get the top x common venues of each borough of a given city and return a dataframe with those venue categoriy ranks. The return the result of the function _return_top_x_venues(df_grouped, num_top_venues)_ you have to input:
* the grouped dataframe discussed earlier
* the number of most common venues you want to extract

As an output you will get a dataframe with the following form:

| Borough | 1st Most Common Venue | 2nd Most Common Venue | 3rd Most Common Venue | ... |
| --- | --- | --- | --- | --- |
| Borough 1 | ... | ... | ... | ... |
| Borough 2 | ... | ... | ... | ... |

In order to finally define our function  _return_top_x_venues_, we will define an auxilary function _return_most_common_venues_:

```python
# Auxilary Function
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Finally define our function _return_top_x_venues_:

```python
def return_top_x_venues(df_grouped, num_top_venues):
    
    # define indicators for numbering 1st, 2nd, 3rd etc.
    indicators = ['st', 'nd', 'rd']
    
    # create columns according to number of top venues
    columns = ['Borough']
    for ind in np.arange(num_top_venues):
        try:
            columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
        except:
            columns.append('{}th Most Common Venue'.format(ind+1))
    
    # create a new dataframe
    neighborhoods_venues_sorted = pd.DataFrame(columns = columns)
    neighborhoods_venues_sorted['Borough'] = df_grouped['Borough']
    
    # write top venues into new dataframe
    for ind in np.arange(df_grouped.shape[0]):
        neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(df_grouped.iloc[ind, :], num_top_venues)
    
    # Return statement
    return neighborhoods_venues_sorted
```

<h5>C. Function that clusteres boroughs of given city by top venues</h5>

This function runs a _k_-means-clustering on a city. That means it labels the boroughs with a cluster-label out of a defined number of clusters and outputs a merged dataframe with all data columns and a new _Cluster Labels_ column

```python
def cluster_boroughs_by_venues(df_city, df_grouped, df_top_venues, k_clusters):
    
    ## 1. Run k-means to cluster the boroughs
    
    # create new dataframe withough 'Borough' column
    df_grouped_clustering = df_grouped.drop('Borough', 1)

    # run k-means clustering
    kmeans = KMeans(n_clusters = k_clusters, random_state = 0).fit(df_grouped_clustering)

    # check cluster labels generated for each row in the dataframe
    #kmeans.labels_[0:10]
    
    ## 2. Create a new dataframe that includes the cluster as well as the top 10 venues for each neighbourhood.

    # add clustering labels
    df_cluster_added = df_top_venues
    df_cluster_added.insert(0, 'Cluster Labels', kmeans.labels_)
    
    df_merged = df_city

    # merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
    df_merged = df_merged.join(df_cluster_added.set_index('Borough'), on = 'Borough')
    
    ## Return Statement
    return(kmeans, df_merged)
```

<h3>5.1 Prepare dataframes for clustering</h3>

Before running the clustering algorithm the two dataframes (_muc_grouped_ and _ffh_grouped_) are prepared and comined to one dataframe, consisting of all borough names and the venue category frequency of occurance.

<h4>1. Remove non similar colums</h4>

As a first step only the venue categories that occure in both cities should be kept and all other venue categories have to be removed. Two new dataframes are created that both have the same column names and therefore same venue categories.

```python
#initiailze list of common columns
comm_col = []

# fill in common columns through for loop
for i in muc_grouped.columns[1:]:
    if i in ffh_grouped.columns[1:]:
        comm_col.append(i)
#  create column list for new dataframes of muc and ffh
muc_col = list(muc_grouped.columns[0:1]) + comm_col
ffh_col = list(ffh_grouped.columns[0:1]) + comm_col

# create new dataframes for muc and ffh
muc_grouped_comm = muc_grouped.loc[:, muc_col]
ffh_grouped_comm = ffh_grouped.loc[:, ffh_col]

# Print statements
print_before = 'Before reomoving non-common columns: \tShape of Munich: {} \t shape of Frankfurt: {}'.format(muc_grouped.shape, ffh_grouped.shape)
print_after = 'After reomoving non-common columns: \tShape of Munich: {} \t shape of Frankfurt: {}' .format(muc_grouped_comm.shape, ffh_grouped_comm.shape)

# Output print
print(print_before)
print(print_after)
```

After cluistering the cluster which contains our current location _Ludwigsvorstadt-Isarvorstadt_ also contains other boroughs of Frankfurt that are similar.

Let's run the _k-means_ algorithm on cities_combined

```python
## 1. Run k-means to cluster the boroughs

# define number of clusters
k_clusters = 8

# create new dataframe withough 'Borough' column
cities_combined_clustering = cities_combined.drop('Borough', axis = 1)

# run k-means clustering
kmeans = KMeans(n_clusters = k_clusters, random_state = 0).fit(cities_combined_clustering)

# check cluster labels generated for each row in the dataframe
#kmeans.labels_[0:10]
    
## 2. Create a new dataframe that includes the cluster as well as the top 10 venues for each neighbourhood.

# add clustering labelsdf_top_venues
cities_cluster_added = cities_combined_clustering
cities_cluster_added.insert(0, 'Cluster Labels', kmeans.labels_)

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
cities_clustered = cities_cluster_added
cities_clustered = cities_clustered.insert(0, 'Borough', cities_combined['Borough'])

## Return Statement   
display(cities_cluster_added)
```
